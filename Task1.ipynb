```
import numpy as np 
import re
import copy

class Maze: 

  def __init__(self):
    self.mazeNum = np.random.randint(0, 9, 200)
    self.newShape = np.array(self.mazeNum).reshape(10, 20)
    self.startPointIndexes = {
        'row': 0,
        'column': 0} 
    self.endPointIndexs = {
        'row': 9,
        'column': 19} 
    self.currentPointIndexes = copy.deepcopy(self.startPointIndexes)
    self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
    self.method1Route = []
    self.method1Route.append(self.currentValue)
    self.method2Route = []
    self.method2Route.append(self.currentValue)
    self.differences = []
    self.endGame = False 
    self.method1Costs = 0
    self.method2Costs = 0
    

  def resetValues(self):
    self.currentPointIndexes = copy.deepcopy(self.startPointIndexes)
    self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
    self.endGame = False

  def showMaze(self):
    print(' '+ re.sub('[\[\]]', '', np.array_str(self.newShape)))  
  
  def startGame(self):
    self.showMaze()
    print('\n')
    print('Route by the search method one: ')
    print(self.searchMethod1())
    self.method1Costs = self.calculateCost(self.method1Route)
    print('TotalCosts in the method one: ', self.method1Costs)
    self.resetValues()
    print('\n')
    print('Route by the search method two (lower differences): ')
    print(self.searchMethod2())
    print('\n')
    print('Differences in the method two:', end='\n')
    print(self.differences)
    self.method2Costs = self.calculateCost(self.differences)
    print('TotalCosts in the method two: ', self.method2Costs)
    print('\n')
    print('Winner Method Cup goes to...:', 'Method One' if self.method1Costs < self.method2Costs else 'Method Two')

  def calculateCost(self, costs):
    totalCosts = 0
    for node in costs:
      totalCosts += node
    return totalCosts

  def searchMethod1(self):
    while self.endGame != True:
      if self.currentPointIndexes['row'] < self.endPointIndexs['row'] and self.currentPointIndexes['column'] < self.endPointIndexs['column']:
        if self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1] < self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]:
          if self.currentPointIndexes['column'] < self.endPointIndexs['column']:
            self.currentPointIndexes['column'] += 1  
            self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
            self.method1Route.append(self.currentValue)
            if self.currentPointIndexes == self.endPointIndexs:
              self.endGame = True
        
          elif self.currentPointIndexes['column'] == self.endPointIndexs['column']:
            if self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] - 1] == 0:
              self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] -1]
              self.method1Route.append(self.currentValue)
              if self.currentPointIndexes == self.endPointIndexs:
                self.endGame = True
            
            elif self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] - 1] != 0:
              if self.currentPointIndexes != self.endPointIndexs:
                self.currentValue = self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]
                self.method1Route.append(self.currentValue)
                if self.currentPointIndexes == self.endPointIndexs:
                  self.endGame = True
            

        elif self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1] > self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]:
          if self.currentPointIndexes['row'] < self.endPointIndexs['row']:
            self.currentPointIndexes['row'] += 1  
            self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
            self.method1Route.append(self.currentValue)
            if self.currentPointIndexes == self.endPointIndexs:
              self.endGame = True
          
          elif self.currentPointIndexes['row'] == self.endPointIndexs['row']:
            if self.newShape[self.currentPointIndexes['row'] -1, self.currentPointIndexes['column']] == 0:
              self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] -1]
              self.method1Route.append(self.currentValue)
              if self.currentPointIndexes == self.endPointIndexs:
                self.endGame = True
       
            elif self.newShape[self.currentPointIndexes['row'] - 1, self.currentPointIndexes['column']] != 0:
              if self.currentPointIndexes != self.endPointIndexs:
                self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
                self.method1Route.append(self.currentValue)
                if self.currentPointIndexes == self.endPointIndexs:
                  self.endGame = True
       
        else:
          if self.currentPointIndexes != self.endPointIndexs:
            self.currentPointIndexes['column'] += 1  
            self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
            self.method1Route.append(self.currentValue)
            if self.currentPointIndexes == self.endPointIndexs:
              self.endGame = True

      elif self.currentPointIndexes['row'] == self.endPointIndexs['row'] and self.currentPointIndexes['column'] < self.endPointIndexs['column']:      
        self.currentPointIndexes['column'] += 1  
        self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
        self.method1Route.append(self.currentValue)
        if self.currentPointIndexes == self.endPointIndexs:
          self.endGame = True

      elif self.currentPointIndexes['row'] < self.endPointIndexs['row'] and self.currentPointIndexes['column'] == self.endPointIndexs['column']:
        self.currentPointIndexes['row'] += 1  
        self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
        self.method1Route.append(self.currentValue)
        if self.currentPointIndexes == self.endPointIndexs:
          self.endGame = True
    
      else: 
        self.endGame = True

    self.resetValues()
    return self.method1Route

  

  def searchMethod2(self):
    while self.endGame != True:
      if self.currentPointIndexes['row'] < self.endPointIndexs['row'] and self.currentPointIndexes['column'] < self.endPointIndexs['column']:
        if abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1]) < abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]):
          self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1]))
          if self.currentPointIndexes['column'] < self.endPointIndexs['column']:
            self.currentPointIndexes['column'] += 1  
            self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
            self.method2Route.append(self.currentValue)
            if self.currentPointIndexes == self.endPointIndexs:
              self.endGame = True
        
          elif self.currentPointIndexes['column'] == self.endPointIndexs['column']:
            if self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] - 1] == 0:
              self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1]))        
              self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] -1]
              self.method2Route.append(self.currentValue)
              if self.currentPointIndexes == self.endPointIndexs:
                self.endGame = True
            
            elif self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] - 1] != 0:
              if self.currentPointIndexes != self.endPointIndexs:
                self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1]))
                self.currentValue = self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]
                self.method2Route.append(self.currentValue)
                if self.currentPointIndexes == self.endPointIndexs:
                  self.endGame = True
            

        elif abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1]) > abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]):
          self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]))
          if self.currentPointIndexes['row'] < self.endPointIndexs['row']:
            self.currentPointIndexes['row'] += 1  
            self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
            self.method2Route.append(self.currentValue)
            if self.currentPointIndexes == self.endPointIndexs:
              self.endGame = True
          
          elif self.currentPointIndexes['row'] == self.endPointIndexs['row']:
            if self.newShape[self.currentPointIndexes['row'] -1, self.currentPointIndexes['column']] == 0:
              self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]))
              self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] -1]
              self.method2Route.append(self.currentValue)
              if self.currentPointIndexes == self.endPointIndexs:
                self.endGame = True
       
            elif self.newShape[self.currentPointIndexes['row'] - 1, self.currentPointIndexes['column']] != 0:
              if self.currentPointIndexes != self.endPointIndexs:
                self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]))
                self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
                self.method2Route.append(self.currentValue)
                if self.currentPointIndexes == self.endPointIndexs:
                  self.endGame = True
       
        else:
          if self.currentPointIndexes != self.endPointIndexs:
            self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] , self.currentPointIndexes['column']+ 1]))
            self.currentPointIndexes['column'] += 1  
            self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
            self.method2Route.append(self.currentValue)
            if self.currentPointIndexes == self.endPointIndexs:
              self.endGame = True

      elif self.currentPointIndexes['row'] == self.endPointIndexs['row'] and self.currentPointIndexes['column'] < self.endPointIndexs['column']:      
        self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column'] + 1]))
        self.currentPointIndexes['column'] += 1  
        self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
        self.method2Route.append(self.currentValue)
        if self.currentPointIndexes == self.endPointIndexs:
          self.endGame = True

      elif self.currentPointIndexes['row'] < self.endPointIndexs['row'] and self.currentPointIndexes['column'] == self.endPointIndexs['column']:
        self.differences.append(abs(self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']] - self.newShape[self.currentPointIndexes['row'] + 1, self.currentPointIndexes['column']]))
        self.currentPointIndexes['row'] += 1  
        self.currentValue = self.newShape[self.currentPointIndexes['row'], self.currentPointIndexes['column']]
        self.method2Route.append(self.currentValue)
        if self.currentPointIndexes == self.endPointIndexs:
          self.endGame = True
    
      else: 
        self.endGame = True

    return self.method2Route
      
```

```

maze = Maze()
maze.startGame()
```
